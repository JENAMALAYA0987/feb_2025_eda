**step-9: drop the column**
- in order to derop a column we need to use **drop** method
- all the netrowk based on your dataframe name similar as string names
- it requires mainly 3 arguments
    - 1.column 
    -2.axixs=0 
     - 3 .inplace
       - once you drop the colum, the dataframe affected
       - the modified dataframe wants to save in a same or differnt name
       - if you want to keep at same name then inplaace= true

```python
name=['raj','abhi','chinna']
age=[20,21,23]
city=['pune','hyd','odisha']
cols=['names','age','city']
id=['a','b','c']
df=pd.DataFrame( zip(name,age,city),index=id,columns=cols)
df
```

```python
df.drop(['age'],axis=1,inplace=True)
df
```

```python
ame=['raj','abhi','chinna']
age=[20,21,23]
city=['pune','hyd','odisha']
cols=['names','age','city']
id=['a','b','c']
df=pd.DataFrame( zip(name,age,city),index=id,columns=cols)
df.drop('a',axis=0,inplace=True)
df
```
