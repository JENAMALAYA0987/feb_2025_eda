**Advantages**
- generally Ml models expect independance between input variables
- there is no relation between the variables
- because of one hot at a time only one will be on, other will off
- this indicates the new columns are independant each other
- this independance behaviour also called as orthogonality
- independant means no relation
- no relation means perpendicukar each other
- perpendicular means 90 degrees phase shift
- 90degrees means ortogonal

**Disadvanatage**
- it will create more columns
- more columns means more data
- more data more complexity w.r.t time and infra
- this is called as curse of dimensionality
